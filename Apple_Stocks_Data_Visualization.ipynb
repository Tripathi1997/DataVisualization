```python
import numpy as np      # for numerical python and other stuffs etc.
import pandas as pd     # for dataframe and other stuffs etc.
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, date, time
```

```python
data = pd.read_excel('apple_stocks.xlsx')  # read data from the file
print(data)
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Day'] = pd.DatetimeIndex(df['Date']).day
print(df)
```

```python
plt.plot(df['Month'], df['High'])
plt.xlabel('Month')
plt.ylabel('Highest Price')
plt.title('Month vs Price (Year - 2017)')
```

```python
plt.plot(df['Month'], df['High'],linestyle='-',marker='',color='r',label='High Price value')
plt.plot(df['Month'], df['Low'],label='low Price Value')

plt.xlabel('Month')
plt.ylabel('Price')
plt.legend()
plt.title('Month vs Price (Year - 2017)')
```

```python
plt.hist(df['Month'])
```

```python
plt.bar(df['Month'],df['High'], color='cyan', width=0.3, label = 'High Price')
plt.bar(df['Month'],df['Low'], color='red', width=0.3, label = 'Low Price')
plt.xlabel('Month')
plt.ylabel('Price')
plt.title('Month vs Price')
plt.legend()
plt.show()
```

```python
# print(len(df))

month_dict = {1:'Jan',2:'Feb',3:'March',4:'April',5:'May',6:'June',7:'July'}
month_high_price_data=[]
month_low_price_data=[]
month_day=[]
month_name=''
def high_low_per_month_plot(month):
#     print('Hello')
    month_name = month_dict.get(month)
    print(month_name)
    
    for i in range(len(df.Month.values)):
        if df.Month.values[i] == month:
            month_high_price_data.append(df.loc[i,'High'])
            month_low_price_data.append(df.loc[i,'Low'])
            month_day.append(df.loc[i,'Day'])
    
    print(month_high_price_data)
    print()
    print(month_low_price_data)
    print()
    print(month_day)

    plt.plot(month_day,month_high_price_data, color='cyan', label = 'High Price',marker='o')
    plt.plot(month_day,month_low_price_data, color='red', label = 'Low Price',marker='*')
    plt.xlabel(f'{month_name} Month')
    plt.ylabel('Price')
    plt.title(' Month vs Price/Day')
    plt.legend()
    plt.show()
```

```python
month = int(input('Enter the month number(1-7): '))
high_low_per_month_plot(month)   # graph plotted for month enterted by user and gets the data
```

```python
month_open_price_data=[]
month_close_price_data=[]
month_day=[]
# month_name=''
def open_close_price_per_month_plot(month):
#     print('Hello')
    month_name = month_dict.get(month)
    print(month_name)
    
    for i in range(len(df.Month.values)):
        if df.Month.values[i] == month:
            month_open_price_data.append(df.loc[i,'Open'])
            month_close_price_data.append(df.loc[i,'Close'])
            month_day.append(df.loc[i,'Day'])
    
    print(month_open_price_data)
    print()
    print(month_close_price_data)
    print()
    print(month_day)

    plt.plot(month_day,month_open_price_data, color='cyan', label = 'Open Price',marker='o')
    plt.plot(month_day,month_close_price_data, color='red', label = 'Close Price',marker='*')
    plt.xlabel(f'{month_name} Month')
    plt.ylabel('Price')
    plt.title(' Month vs Price/Day')
    plt.legend()
    plt.show()
```
